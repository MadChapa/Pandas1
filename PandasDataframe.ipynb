Column Addition:

```python
import pandas as pd
info={
    'one':pd.Series([1,2,3,4,5,6], index=['a','b','c','d','e','f']),
    'two':pd.Series([1,2,3,4,5,6,7,8],index=['a','b','c','d','e','f','g','h'])}
df=pd.DataFrame(info)
```

```python
df['three']=pd.Series([20,40,60],index=['a','b','c'])
print(df)
```

```python
df['four']=df['one']+df['three']
print(df)
```

Column Deletion:

```python
import pandas as pd
info = {'one' : pd.Series([1, 2], index= ['a', 'b']),
        'two' : pd.Series([1, 2, 3], index=['a', 'b', 'c'])}  
df=pd.DataFrame(info)
print("The DataFrame:")
print(df)
```

```python
#delete column using del functiom
print("Delete the first column:")
del df['one']
print(df)
```

```python
#delete column using pop function
print("delete the second column")
df=pd.DataFrame(info)

df.pop('two')
print(df)
```

 Row Selection Addition and Deletion:

```python
#Selection by label
import pandas as pd
info={
    'one':pd.Series([1,2,3,4,5],index=['a','b','c','d','e']),
    'two':pd.Series([1,2,3,4,5,6],index=['a','b','c','d','e','f'])}
df=pd.DataFrame(info)
print(df.loc['b'])
```

```python
#Selection by integer location
print (df.iloc[3])  #a[3] ko index mah j value xa tahi value aauxa
```

```python
#Slice Rows

print(df[2:5])
```

Addition of rows

```python
#using append function
import pandas as pd
d=pd.DataFrame([[7,8],[9,10]],columns=['x','y'])
d2=pd.DataFrame([[11,12],[13,14]],columns=['x','y'])
d=d.append(d2)
print(d)
```
