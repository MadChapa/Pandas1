```python
#Aggregation
import pandas as pd  
import numpy as np  
dic = {'Name': ['San', 'Kes', 'Susil', 'Bamm'],  
   'Percentage': [82, 98, 91, 97],  
   'Course': ['BSC','BEd','M.Phill','BA']} 
df=pd.DataFrame(dic)
df
```

```python
grouped=df.groupby('Course')
print(grouped['Percentage'].agg(np.mean))
```
