```python
Creating Series from Array:
```

```python
import pandas as pd  
import numpy as np  
info = np.array(['P','a','n','d','a','s'])  
a = pd.Series(info)  
print(a)   
```

```python
Create a series from dict:
```

```python
#import the pandas library   
import pandas as pd  
import numpy as np  
info = {'x' : 0., 'y' : 1., 'z' : 2.}  
a = pd.Series(info)  
print (a)  
```

Creatinng a series using Sclar:

```python
import pandas as pd  
import numpy as np  
x = pd.Series(4, index=[0, 1, 2, 3])  
print (x)  
```

Pandas Series.map() function:

```python
import pandas as pd  
import numpy as np  
a = pd.Series(['Java', 'C', 'C++', np.nan])  #np.nan: not a number
a.map({'Java': 'Core'})  
```

```python
  
a = pd.Series(['Java', 'C', 'C++', np.nan])  
a.map({'Java': 'Core'})  
a.map('I like {}'.format, na_action='ignore')  
```

Pandas Series.std() Function:

```python
import pandas as pd  
import numpy as np  
   
#Create a DataFrame  
info = {  
    'Name':['san','chap','ke','pou'],  
   'sub1_Marks':[52,38,42,37],  
   'sub2_Marks':[41,35,29,36]}   
data = pd.DataFrame(info)  
data  
# standard deviation of the dataframe  
data.std() 
```

Pandas Series.to_frame()

```python
import pandas as pd   
import matplotlib.pyplot as plt     
emp = ['Parker', 'John', 'Smith', 'William']   
id = [102, 107, 109, 114]     
emp_series = pd.Series(emp)   
id_series = pd.Series(id)     
frame = { 'Emp': emp_series, 'ID': id_series }   
result = pd.DataFrame(frame)     
print(result)  
```

Pandas Series.value_counts():


```python
import pandas as pd 
import numpy as np
index = pd.Index([1, 3, 2, 2, 1, np.nan])  
index.value_counts()  
a = pd.Series([1, 3, 2, 2, 1, np.nan])  
a.value_counts(normalize=True)  
```

```python
a.value_counts(bins=2)  

```

```python
a.value_counts(dropna=False)  
```
